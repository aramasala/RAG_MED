This notebook shows how to use RAG_MED from Python: clean a PDF and generate QA pairs.

```python
from pathlib import Path
from rag_med import clean_pdf, generate_qa_from_pdf
```

## Clean PDF

Remove bibliography and appendix sections from a medical PDF (paths are examples).

```python
input_pdf = Path("document.pdf")
output_pdf = Path("document_cleaned.pdf")
if input_pdf.exists():
    ok = clean_pdf(input_pdf, output_pdf)
    print(f"Clean: {ok}")
else:
    print("Place a PDF as document.pdf in the project root to run this cell.")
```

## Generate QA from PDF

```python
pdf_path = Path("document.pdf")
if pdf_path.exists():
    results = generate_qa_from_pdf(pdf_path, Path("qa_result.json"), num_questions=3)
    for r in results:
        print(f"Q: {r.question[:80]}...")
        print(f"A: {r.answer[:80]}...")
else:
    print("Place a PDF as document.pdf to run this cell.")
```

Configuration is loaded from `.env` and the `configs` package (see project root).
