# RAG_MED — Usage demo

This notebook shows how to use RAG_MED from Python: clean a PDF and generate QA pairs.

```python
from pathlib import Path
from rag_med import clean_pdf, generate_qa_from_pdf
```

## Clean PDF

Remove bibliography and appendix sections from a medical PDF (paths are examples).

```python
input_pdf = Path("document.pdf")
output_pdf = Path("document_cleaned.pdf")
if input_pdf.exists():
    ok = clean_pdf(input_pdf, output_pdf)
    print(f"Clean: {ok}")
else:
    print("Place a PDF as document.pdf in the project root to run this cell.")
```

## Generate QA from PDF

```python
pdf_path = Path("document.pdf")
output_file = Path("qa_result.json")
if pdf_path.exists():
    # Set evaluate_with_valueai=True to run RAGAS + ValueAI RAG evaluation (requires VALUEAI_* in .env)
    results = generate_qa_from_pdf(
        pdf_path, output_file, num_questions=3,
        evaluate_with_valueai=False,
        summary_file=None,  # or Path("qa_result_valueai_eval.json") when using --valueai-eval
    )
    for r in results:
        print(f"Q: {r.question[:80]}...")
        print(f"A: {r.answer[:80]}...")
else:
    print("Place a PDF as document.pdf to run this cell.")
```

Configuration is loaded from `.env` and the `configs` package (project root). Set ValueAI credentials (`VALUEAI_BASE_URL`, `VALUEAI_USERNAME`, `VALUEAI_PASSWORD`) and optionally `METRICS_LLM_MODEL_NAME` for QA generation and evaluation.
